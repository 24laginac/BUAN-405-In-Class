```python
from sklearn import cluster
import pandas as pd

# replace the filename with the path to the dataset you want to use
all_data = pd.read_excel('/content/testforclusternew.xlsx')
num_data = all_data.iloc[:, 2:5]
k_clusters = 2   # 2 or 3 is normally a good number of clusters
labels = cluster.KMeans(n_clusters = k_clusters).fit(num_data).labels_
all_data['Labels'] = labels
print (all_data)
```

```python
cont_data = all_data.iloc[:, 2:5]
cont_data
```
